```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import json
```

```python
page = requests.get('https://en.wikipedia.org/wiki/List_of_video_game_developers').content
```

```python
soup = bs(page)
```

```python
tables = soup.findAll("table", {'class':['wikitable']})
```

```python
table = tables[1]
```

```python
table
```

```python
columns = [i.text.strip() for i in table.findAll("th")]
```

```python
columns
```

```python
rows = table.findAll('tr')
```

```python
len(rows)
```

```python
t = rows[1]
```

```python
[i.text for i in t.findAll("td")]
```

```python
rows[1]
```

```python
content = []
for row in rows[1:]:
    r = [i.text.strip() for i in row.findAll("td")]
    content += [
        {
            'Developer': r[0],
             'City': r[1],
             'Administrative division': r[2],
             'Country': r[3],
             'Est.': r[4],
             'Notable games, series or franchises': [i.text.strip() for i in row.findAll('i')],
             'Notes': r[6] if len(r) == 7 else None
        } 
    ]
```

```python
content
```

```python
len(content)
```

```python
df = pd.DataFrame(content)
```

```python
with open("Developers.json", 'w') as f:
    json.dump(content, f)
```
